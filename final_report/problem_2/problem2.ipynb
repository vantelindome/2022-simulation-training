```python
%matplotlib inline
import math
import matplotlib
%config InlineBackend.figure_format = 'retina'
import matplotlib.cm as cm  # colormap
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
```

```python
# (2)

plt.rcParams["text.usetex"] =True
plt.rcParams["font.size"] = 30

from matplotlib.collections import PatchCollection
from matplotlib.patches import Circle

resolution = 100 # the number of vertices
Np=1000
L=50.0

patches = [[] for i in range(4)]  #空の二重配列を定義

fig = plt.figure(figsize=(8,8))

for j in range(1):
    ax = fig.add_subplot(111)
    x, y, a = np.loadtxt("./2-2/coord_T5.000_1_problem2-2.dat", comments='#', unpack=True, usecols=[0,1,4])
    for i in range(Np):
        circle = mpatches.Ellipse((x[i],y[i]), a[i], a[i])
        patches[j-1].append(circle)
    plt.xlim(0, L)
    plt.ylim(0, L)    
    colors = a
    p = PatchCollection(patches[j-1], cmap=matplotlib.cm.jet, alpha=1.0,ec='k')
    p.set_array(colors)
    p.set_clim(0.8,1.2)
    ax.add_collection(p)
    ax.spines['top'].set_linewidth(2)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    ax.spines['right'].set_linewidth(2)
    plt.tick_params(which='major',width = 1, length = 10)
    plt.tick_params(which='minor',width = 1, length = 5)
 
plt.xlabel(r"$x/a$",color='k', size=35)
plt.ylabel(r"$y/a$",color='k', size=35)
plt.title("$T^*=5.0$, $t=10t_0$")

plt.subplots_adjust(wspace=0.2, hspace=0.25)
plt.savefig('./2-2/problem2-2.pdf',bbox_inches="tight")
plt.show()
```

```python
# (5)

plt.rcParams["text.usetex"] =True
plt.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'

fig = plt.figure(figsize=(12, 6))
temp=['K/Np','U/Np','(K+U)/Np']

plt.subplot(111)
ax=fig.add_subplot(111)
for i in range(0,3):
    t, energy = np.loadtxt('./2-4/energy.dat', unpack=True, usecols=[0,i+1])
    plt.plot(t, energy, marker='.', markersize=3, linewidth=2, label="${:.}$".format(temp[i]))


plt.xlabel('$t/t_0$')
plt.ylabel('Energy')

ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
plt.legend()

plt.savefig('./2-4/problem2-4.png')
plt.savefig('./2-4/problem2-4.pdf')
plt.show()
```
