```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.cm as cm  # colormap
import numpy as np
import math
```

```python
plt.rcParams["text.usetex"] =True 

fig = plt.figure(figsize=(8, 4))
temp=[0.1,1.0,10]

ax1=fig.add_subplot(111)
for i in range(0,3):
    x, y = np.loadtxt('./1-4/m={:.1f}.txt'.format(temp[i]), unpack=True, usecols=[1,2])
    plt.plot(x, y, marker='.', label=r"$m^*={:.1f}$".format(temp[i]))


plt.title("The trajectry of a particle based on Langevin eqation", fontsize=20)
plt.xlabel('x')
plt.ylabel('y')
ax1.spines['top'].set_linewidth(2)
ax1.spines['bottom'].set_linewidth(2)
ax1.spines['left'].set_linewidth(2)
ax1.spines['right'].set_linewidth(2)

plt.legend()
plt.savefig('./1-4/problem1-4.png')
plt.savefig('./1-4/problem1-4.pdf')
plt.show()
```

```python
plt.rcParams["text.usetex"] =True 

fig = plt.figure(figsize=(8, 4))
temp=[0.1,1.0,10]

ax1=fig.add_subplot(111)
for i in range(0,3):
    t, r = np.loadtxt('1-4/r2_m{:.1f}.dat'.format(temp[i]), unpack=True, usecols=[0,3])
    plt.plot(t, r, '.', label=r"$m^*={:.1f}$".format(temp[i]))

    t= np.linspace(0.01, 100, 10000) 
    f= 4*temp[i]*(t/temp[i] + np.exp(-t/temp[i]) -1)
    plt.plot(t,f, "--",markersize=3,linewidth = 2.0, color="k", label=r"$m^*={:.1f}$ theory".format(temp[i]))


plt.title("", fontsize=20)
plt.xlabel('$t/t_B$')
plt.ylabel('$\langle \Delta \bm{r}^2 \rangle /a^2$')
ax1.spines['top'].set_linewidth(2)
ax1.spines['bottom'].set_linewidth(2)
ax1.spines['left'].set_linewidth(2)
ax1.spines['right'].set_linewidth(2)

plt.legend()
plt.savefig('1-5/problem1-5.png')
plt.savefig('1-5/problem1-5.pdf')
plt.show()
```
