```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.cm as cm  # colormap
import numpy as np
import math
```

```python
# (4)

plt.rcParams["text.usetex"] =True 

fig = plt.figure(figsize=(12, 6))
temp=[0.1,1.0,10]

for i in range(0,3):
    ax=fig.add_subplot(1,3,i+1)
    t, x = np.loadtxt('./1-4/m={:.1f}.txt'.format(temp[i]), unpack=True, usecols=[0,1])
    plt.plot(t, x, marker='.', markersize=2, linewidth=1, label="$x$")

    t, y = np.loadtxt('./1-4/m={:.1f}.txt'.format(temp[i]), unpack=True, usecols=[0,2])
    plt.plot(t, y, marker='.', markersize=2, linewidth=1, label="$y$")
    plt.title("$m^*={:.1f}$".format(temp[i]), fontsize=20)
    plt.xlabel('$t/t_0$')
    plt.ylabel('$x/a$,$y/a$')
    ax.spines['top'].set_linewidth(2)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    ax.spines['right'].set_linewidth(2)
    plt.legend()

plt.savefig('./1-4/problem1-4.png')
plt.savefig('./1-4/problem1-4.pdf')
plt.show()
```

```python
# (5)

plt.rcParams["text.usetex"] =True
plt.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'

fig = plt.figure(figsize=(12, 6))
temp=[0.1,1.0,10]

plt.subplot(111)
ax=fig.add_subplot(111)
for i in range(0,3):
    t, r = np.loadtxt('./1-5/r2_m{:.1f}.dat'.format(temp[i]), unpack=True, usecols=[0,1])
    plt.plot(t, r, 'o', markersize=3, label="$m={:.1f}$".format(temp[i]))

    t = np.linspace(0.01, 100, 10000)
    f = 4 * temp[i] * (t/temp[i] + np.exp(-t/temp[i]) -1)
    plt.plot(t, f, '--', markersize=2, linewidth=1, color="k", label="$m={:.1f}$-theory".format(temp[i]))


plt.xlabel('$t/t_0$')
plt.ylabel('$\langle \mathbf{r}(t) \rangle /a^2$')
plt.xscale('log')
plt.yscale('log')

ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
plt.legend()

plt.savefig('./1-5/problem1-5.png')
plt.savefig('./1-5/problem1-5.pdf')
plt.show()
```
